```python
import plotly.plotly as py
import pandas as pd
import plotly
```

```python
df_all = pd.read_csv("../CSVs/table_second_step_merge_signatures_coordinates.csv")
```

```python
df_platforms = df_all.query("platform_opendata == platform_opendata")
```

```python
df_platforms
```

```python
# aqui mostra a quantidade de url assinadas em cada pais
df_gru = df_platforms["url"].groupby([df_platforms["country"],df_platforms["latitude"],df_platforms["longitude"],df_platforms["platform_opendata"]],sort=True).count().sort_values(ascending=False)
df_gru.to_csv("groups_country.csv")
```

```python
df_plotar = pd.read_csv("groups_country.csv",names=["country","latitude","longitude","platform_opendata","count"])
```

```python
df_plotar
```

```python
df_plotar['text'] = df_plotar['country'] + '<br>URLs que respondem a assinatura ' + (df_plotar['platform_opendata']) + '<br>Quantidade '+(df_plotar['count']).astype(str)
limits = [(0,5),(5,10),(10,20),(20,50),(50,2500)]
#limits = [(0,2),(3,10),(11,20),(21,50),(50,3000)]
colors = ["rgb(0,116,217)","rgb(255,65,54)","rgb(133,20,75)","rgb(255,133,27)","lightgrey"]
cities = []
scale = 5

for i in range(len(limits)):
    lim = limits[i]
    df_sub = df_plotar[lim[0]:lim[1]]
    city = dict(
        type = 'scattergeo',
        #locationmode = 'USA-states',
        lon = df_sub['longitude'],
        lat = df_sub['latitude'],
        text = df_sub['text'],
        marker = dict(
            size = 10,#*scale,
            # sizeref = 2. * max(df_sub['pop']/scale) / (25 ** 2),
            color = colors[i],
            line = dict(width=0.5, color='rgb(40,40,40)'),
            sizemode = 'area'
        ),
        name = '{0} - {1}'.format(lim[0],lim[1]) )
    cities.append(city)

layout = dict(
        title = 'Quantidade de URLs que respondem a alguma assinatura',
        showlegend = True,
        geo = dict(
            scope='world',
            projection=dict( type = 'natural earth'),
            showland = True,
            landcolor = 'rgb(217, 217, 217)',
            subunitwidth=1,
            countrywidth=1,
            subunitcolor="rgb(255, 255, 255)",
            countrycolor="rgb(255, 255, 255)"
        ),
    )

fig = dict(data=cities, layout=layout)
py.iplot(fig, validate=False, filename='URLs que utilizam alguma plataforma')
```
