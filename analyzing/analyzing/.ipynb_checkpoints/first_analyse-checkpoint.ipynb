```python
def filter_initials(line):
    initials= ""
    start_gov = line.find("gov")
    if(start_gov != -1):
        if(line[start_gov+3:start_gov+4] == "."):
            start_initials = line.find(".",start_gov+3)
            initials= line[start_initials+1:start_initials+3]

            if(initials == "uk"):
                initials= "gb"

        elif(line[start_gov+3:start_gov+4] == "/"):
            initials= "us"

        return initials
    else:
        start_gov = line.find(".ca/")
        initials= "ca"
        return initials
```

```python
# crio um dataframe do csv com as url que contem data, open ou transp como parte da url
df_gov = pd.read_csv('../CSVs/gov2.csv', delimiter=',', names=['id','url','date_catch','index_fraction'])

# tiro as duplicatas para deixar somente url 
df_gov.drop_duplicates(['url'], keep='last', inplace=True)
```

```python
# use o csv de siglas e paises para criar um dicionario
with open('../CSVs/list_initials.csv') as file_map_csv:
    dict_initials_country = {}
    initials_country = csv.DictReader(file_map_csv)
    for row in initials_country:
        dict_initials_country[row['sigla'].lower()] = row['pais']
        print(row['sigla'], row['pais'])
```

```python
# criei uma nova coluna e acrecentei a ela ainitialsretirada respectivamente da url de cada linha
df_gov['initials'] = df_gov['url'].apply(filter_initials)
```

```python
# mapiando a serieinitialscom a serie pais usando a função dict_sigla_pais
df_gov['country'] = mapping.map(dict_initials_country)
```

```python
def asignature(url):
   
    import datetime
    import urllib.request
    import ssl
    import json
    import pandas as pd

    ctx = ssl.create_default_context()
    ctx.check_hostname = False
    ctx.verify_mode = ssl.CERT_NONE

    #urls = pd.read_csv("gov_filtrado_open_data_transp.csv", delimiter=',', encoding='latin-1')
   # print(len(urls))
    #print('start: ' + str(datetime.datetime.now()))

    
    #configure products signature
    ckan_sig = '/api/action/site_read'
    socrata_sig = '/api/catalog/v1'
    arcgis_sig = '/api/v2'
    odsoft_sig = '/api/v2'
    catalog = None

    ckan_error = False
    socrata_error = False
    arcgis_error = False
    odsoft_error = False

        #get redirected domain to verify duplicated URLs

    #url_id = str(row['id'])
    #url = str(row['url'])

    try:
        req = urllib.request.Request(
            url, 
            data=None, 
            headers={
                'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'
            }
        )
        resp = urllib.request.urlopen(req, timeout=20, context=ctx)
    except:
        #print(url_id + ';' + url + ';;ERROR')
        #print("dddddd")
        pass
    else:
        domain = resp.geturl()

        #CKAN
        try:
            req = urllib.request.Request(
                url + ckan_sig, 
                data=None, 
                headers={
                    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'
                }
            )
            resp = urllib.request.urlopen(req, timeout=20, context=ctx)
        except Exception as e: #raises HTTP, URL and refused connection errors
            ckan_error = True
        else:
            try:
                response_dict = json.loads(resp.read())
            except ValueError: #raises JSON format error of a non-expected result (API not found)
                pass
            else:
                if 'success' in response_dict or 'help' in response_dict or 'result' in response_dict:
                    catalog = 'CKAN'

        #SOCRATA
        try:
            req = urllib.request.Request(
                url + socrata_sig, 
                data=None, 
                headers={
                    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'
                }
            )
            resp = urllib.request.urlopen(req, timeout=20, context=ctx)
        except: #raises HTTP, URL and refused connection errors
            socrata_error = True
        else:
            try:
                response_dict = json.loads(resp.read())
            except ValueError: #raises JSON format error of a non-expected result (API not found)
                pass
            else:
                if 'results' in response_dict:
                    catalog = 'SOCRATA'

        #ARCGIS
        try:
            req = urllib.request.Request(
                url + arcgis_sig, 
                data=None, 
                headers={
                    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'
                }
            )
            resp = urllib.request.urlopen(req, timeout=20, context=ctx)
        except: #raises HTTP, URL and refused connection errors
            arcgis_error = True
        else:
            try:
                response_dict = json.loads(resp.read())
            except ValueError: #raises JSON format error of a non-expected result (API not found)
                pass
            else:
                if 'datasets' in response_dict and domain.find('hub.arcgis.com') == -1:
                    catalog = 'ARCGIS'

        #OPENDATASOFT
        try:
            req = urllib.request.Request(
                url + odsoft_sig, 
                data=None,  
                headers={
                    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'
                }
            )
            resp = urllib.request.urlopen(req, timeout=20, context=ctx)
        except: #raises HTTP, URL and refused connection errors
            odsoft_error = True
        else:
            try:
                response_dict = json.loads(resp.read())
            except ValueError: #raises JSON format error of a non-expected result (API not found)
                print("ww")
                pass
            else:
                if 'links' in response_dict:
                    catalog = 'OPENDATASOFT'

        if ckan_error and socrata_error and arcgis_error and odsoft_error:
            #print(url_id + ';' + url + ';' + domain + ';none')
            print("aaaaaaaa")
        else:
            #print(url_id + ';' + url + ';' + domain + ';' + catalog)
            print("\nbbbbbbbbb\n")
            return catalog
```
