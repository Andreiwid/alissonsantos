```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
site = requests.get("https://www.realifewebdesigns.com/web-marketing/abbreviations-countries.asp")
```

```python
soup = BeautifulSoup(site.content,"lxml")
```

```python
lista = soup.findAll("li")
lista_v2 = lista[8:-12]
```

```python
df_lista = pd.DataFrame(lista_v2,columns=['tag_li'])
```

```python
df_lista.to_csv("../CSVs/lista_siglas_ingles.csv")
```

```python
df_nova_lista = pd.read_csv("../CSVs/lista_siglas_ingles.csv")
```

```python
df_nova_lista
```

```python
def get_sigla(line):
    inicio = line.find(">")
    sigla = line.find("=",inicio+1)
    palavra = line[inicio+1:sigla].strip()
    return palavra.lower().strip()
```

```python
def get_name_country(line):
    inicio = line.find("=")
    fim = line.find("<",inicio+1)
    palavra = line[inicio+1 : fim]
    return palavra.lower().strip()
```

```python
 df_nova_lista['sigla'] = df_nova_lista['tag_li'].apply(get_sigla)
```

```python
df_nova_lista['pais'] = df_nova_lista["tag_li"].apply(get_name_country)
```

```python
df_nova_lista.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
df_nova_lista.drop("tag_li", axis=1, inplace=True)
```
