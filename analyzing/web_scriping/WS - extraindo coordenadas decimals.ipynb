```python
def separa_sigla(line):
    indice_inicio_sigla = line.find("td>")
    indice_fim_sigla = line.find("</td")
    sigla = line[indice_inicio_sigla + 3:indice_fim_sigla]
    print(sigla.lower())
    return sigla.lower()
```

```python
def separa_latitude(line):
    indice_inicio_latitude = line.find("td>",15)
    indice_fim_latitude = line.find("<td>",indice_inicio_latitude)
    #print(indice_fim_latitude)
    latitude = line[indice_inicio_latitude + 3: indice_fim_latitude - 6]
    print(latitude)
    return latitude
```

```python
def separa_longitude(line):
    indice_inicio_longitude = line.find("td>",36)
    indice_fim_longitude = line.find("<td>",indice_inicio_longitude)
    #print(indice_fim_longitude)
    longitude = line[indice_inicio_longitude + 3: indice_fim_longitude -6]
    print(longitude)
    return longitude
```
